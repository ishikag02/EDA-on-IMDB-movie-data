```python
#20. Classify Movies Based on Ratings [Excellent, Good, and Average]
df.columns
```

```python
def rating(rating):
    if rating>=8.0:
        return "Excellent"
    elif rating>=6.0:
        return "Good"
    else:
        return "Average"
```

```python
df['rating_cat']=df['Rating'].apply(rating)
```

```python
#22. Find Unique Values From Genre 
df['Genre']
```

```python
list1=[]
for value in df['Genre']:
    list1.append(value.split(','))
list1
```

```python
one_d=[]
for item in list1:
    for item1 in item:
        one_d.append(item1)
one_d
```

```python
uni_list=[]
for item in one_d:
    if item not in uni_list:
        uni_list.append(item)
uni_list
    
```
